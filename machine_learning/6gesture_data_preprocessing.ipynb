```python
import collections
import os
import skimage
from skimage.io import imread
from skimage.transform import resize
import scipy
import numpy as np

def reformat(path):
    videos = collections.defaultdict(list)
#     subject = []
#     gesture_class = []
#     index = []
#     pi_mark = []
    count = 0
    for root, dirs, files in os.walk(path):
        jpg_files = [x for x in files if x.endswith('.jpg') and not x.startswith('._')]
        print(root)
        if len(jpg_files) > 0:
            path_split = root.split(os.path.sep)
            subject = path_split[-4]
            frames = []
            if subject.endswith('free_log'):
                subject = subject[:-9]
                gesture_class = '6'
                pi_mark = path_split[-1]
                
                for fn in sorted(jpg_files):
                    img = imread(os.path.join(root, fn), as_gray=True)
                    img = skimage.transform.resize(img, (48, 64))
                    frames.append(img)
                    
                for index in range(len(jpg_files)-16):
                    sample = np.stack(frames[index:index+16], axis=2)
                    videos['_'.join([subject, gesture_class, str(index), pi_mark])].append(sample)
                    
            else:
                gesture_class = path_split[-3] 
                subject = subject[:-4]
                index = path_split[-2]
                pi_mark = path_split[-1]
                
                for fn in sorted(jpg_files):
                    img = imread(os.path.join(root, fn), as_gray=True)
                    img = resize(img, (48, 64))
                    frames.append(img)
                np_frames = np.stack(frames, axis=2)
                resample = scipy.ndimage.zoom(np_frames, (1,1,16/len(frames)))
                videos['_'.join([subject, gesture_class, index, pi_mark])].append(resample)
                
#             print(subject, gesture_class, index, pi_mark, len(jpg_files))
    return videos
```

```python
videos = reformat('/Volumes/BACKUP_DISK/grw/new/data-part1')
print(videos.keys())
```

```python
#process IMU data
IMU_path = '/Volumes/BACKUP_DISK/grw/new/data-part1/IMU.txt'

# read time and IMU of txt
with open(IMU_path, 'r') as f:
    _ = f.readline()
    content = f.readlines()
    time = []
    temp_data = []
    
    for i, line in enumerate(content):
        line = line.strip().split()
        if len(line) == 7 and line[0] != 'test':
            try:
                time.append(int(line[0]))
                temp_data.append(list(map(float,line[1:4]))+list(map(int,line[4:7])))
            except:
                print(i, line)
```

```python
#scale data
from sklearn import preprocessing
temp_data = np.array(temp_data)
temp_mean = temp_data.mean(axis=0)
temp_std = temp_data.std(axis=0)
print('temp_mean', temp_mean) 
print('temp_std', temp_std)
scale_data = temp_data - temp_mean
scale_data = scale_data/temp_std
print(scale_data.mean(axis=0))
print(scale_data.std(axis=0))
```

```python
# format data according to csv
import os
import bisect
data = []
labels = []
subjects = []
index = []
csv_dir = '/Volumes/BACKUP_DISK/grw/new/csv-part1'
for fn in os.listdir(csv_dir):
    if fn.endswith('log.csv'):
        temp = fn.split('_')
        subject = temp[0]
        with open(os.path.join(csv_dir, fn),'r') as f:
            content = f.readlines()
            
        if temp[1] == 'free' and len(content) >= 2:
            label = 6
            l = content[1].split(',')
            st,en = int(l[1]),int(l[2])
            index_st = bisect.bisect_left(time, st)
            index_en = bisect.bisect_right(time, en)
            data.append(temp_data[index_st:index_en])
            labels.append(int(label))
            subjects.append(subject)
            index.append(0)
        else:
            for l in content[1:]:
                l = l.split(',')
                label = l[0]
                st,en = int(l[1]), int(l[2])
                index_st = bisect.bisect_left(time, st)
                index_en = bisect.bisect_right(time, en)
                data.append(temp_data[index_st:index_en])
                labels.append(int(label))
                subjects.append(subject)
                if len(labels) == 1 or labels[-1] != labels[-2]:
                    index.append(0)
                else:
                    index.append(index[-1]+1)
                
print(len(data))
print(len(labels))
print(len(subjects))
print(len(index))
```
