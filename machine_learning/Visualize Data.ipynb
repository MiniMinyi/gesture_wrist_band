```python
import numpy as np
import matplotlib.pyplot as plt
path = 'video_array_raw/'
test = np.load(path+'2.npy')
print(test.shape)
```

```python
from skimage.transform import rescale

for frame in range(test.shape[0]):
    rescale_image = rescale(test[frame,:,:], 0.5)
    print(rescale_image.shape)
    plt.imshow(rescale_image)
    plt.show()
```
